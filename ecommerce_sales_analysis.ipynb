```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ecommerce_sales.csv', encoding='latin1')
df.head()
```

```python
import matplotlib.pyplot as plt

profit_by_category = df.groupby('Category')['Profit'].sum().sort_values()

profit_by_category.plot(kind='bar', title='Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
