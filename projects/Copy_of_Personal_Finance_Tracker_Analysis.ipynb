```python
import pandas as pd

data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May"],
    "Income": [3200, 3400, 3100, 3600, 3800],
    "Expenses": [2100, 2300, 2000, 2500, 2400],
    "Rent": [1200, 1200, 1200, 1200, 1200],
    "Food": [400, 450, 380, 500, 420],
    "Transport": [200, 180, 220, 210, 190]
}

df = pd.DataFrame(data)
df
df["Savings"] = df["Income"] - df["Expenses"]
df
import matplotlib.pyplot as plt

plt.plot(df["Month"], df["Income"], marker="o", label="Income")
plt.plot(df["Month"], df["Expenses"], marker="o", label="Expenses")
plt.plot(df["Month"], df["Savings"], marker="o", label="Savings")

plt.title("Personal Finance Tracker Overview")
plt.legend()
plt.show()
```
